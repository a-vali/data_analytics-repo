```python
with open("./Data/groceries.csv") as fp:
    groceries_baskets=fp.readlines()
groceries_baskets=[set(basket[:-1].split(',')) for basket in groceries_baskets]
```

```python
groceries_baskets[:5]
```

```python
from collections import defaultdict
from itertools import permutations
```

```python
# support(A):
# find the number of A unique occurance in baskets divided by total number of baskets
# probability of buying item A:

item_count=defaultdict(int)
for basket in groceries_baskets:
    for item in set(basket):
        item_count[item]+=1
print(item_count)

# calculate the probabbility of every item support(A):
# total number of unique items:
total=0
for basket in groceries_baskets:
    total+=len(basket)
    
print(total)

item_probability={}
for item in item_count:
    item_probability[item]=item_count[item]/total
print(item_probability)
```

```python
# support(unioin(A,B))
pair_count=defaultdict(int)
for basket in groceries_baskets:
    for pair in permutations(basket,2):
        pair_count[pair]+=1
print(pair_count)

pair_probability={}
for pair in pair_count:
    pair_probability[pair]=pair_count[pair]/total

print(pair_probability)
```

```python
# find confid(A=>B)
# confid(A=>B)=support(union(A,B))/support(A)

rules={}
for pair,value in pair_count.items():
    rules[pair]=value/item_count[pair[0]]
    
print([f"{rule} = > {rules[rule]:.3f}" for rule in rules if rules[rule]>=0.6])
```
